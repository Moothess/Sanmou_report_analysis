```python
import win32api
import win32con
import win32gui


def check_dpi_issues():
    """检查DPI缩放问题"""
    # 获取系统DPI设置
    try:
        # 获取主显示器的DPI
        hdc = win32gui.GetDC(0)
        dpi_x = win32api.GetDeviceCaps(hdc, win32con.LOGPIXELSX)
        dpi_y = win32api.GetDeviceCaps(hdc, win32con.LOGPIXELSY)
        win32gui.ReleaseDC(0, hdc)

        standard_dpi = 96  # 100%缩放的DPI
        scale_x = dpi_x / standard_dpi
        scale_y = dpi_y / standard_dpi

        print(f"DPI设置: X={dpi_x}, Y={dpi_y}")
        print(f"缩放比例: X={scale_x:.2f}, Y={scale_y:.2f}")

        if scale_x != 1.0 or scale_y != 1.0:
            print("⚠️ DPI缩放可能导致坐标不匹配")
            return True
    except Exception as e:
        print(f"无法获取DPI信息: {e}")

    return False


# 检查DPI问题
has_dpi_issue = check_dpi_issues()
```
